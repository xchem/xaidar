```python
import os
def getFilesList( path ):
    """
    Return the list of file names that exist in input directory
    Args
    - path ( pathlist.Path() object)
    """
    list = [ file for file in os.listdir( path) if os.path.isfile( path / file) ]
    list.sort()
    return list 
```
